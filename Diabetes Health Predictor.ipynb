```python
education_categories = {
    1: 'No School', 2: 'Primary School', 3: 'Secondary School',
    4: 'Graduate', 5: 'Post Graduate', 6: 'Ph.D'
}

income_categories = {
    1: 'Poor', 2: 'Poor', 3: 'Poor',
    4: 'Moderate', 5: 'Moderate', 6: 'Moderate',
    7: 'Rich', 8: 'Rich'
}

dataset['Education_Category'] = dataset['Education'].map(education_categories)
dataset['Income_Category'] = dataset['Income'].map(income_categories)

education_order = ['No School', 'Primary School', 'Secondary School', 'Graduate', 'Post Graduate', 'Ph.D']
income_order = ['Poor', 'Moderate', 'Rich']

dataset['Education_Category'] = pd.Categorical(dataset['Education_Category'], categories=education_order, ordered=True)
dataset['Income_Category'] = pd.Categorical(dataset['Income_Category'], categories=income_order, ordered=True)

cross_tab_education_diabetes = pd.crosstab(index=dataset['Education_Category'], columns=dataset['Diabetes_012'], normalize='index') * 100
cross_tab_income_diabetes = pd.crosstab(index=dataset['Income_Category'], columns=dataset['Diabetes_012'], normalize='index') * 100

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

sn.heatmap(cross_tab_education_diabetes, annot=True, cmap='viridis', fmt='.2f', cbar=True, ax=axes[0])
axes[0].set_title('Education_Category vs. Diabetes_012')
axes[0].set_xlabel('Diabetes_012')
axes[0].set_ylabel('Education Category')

sn.heatmap(cross_tab_income_diabetes, annot=True, cmap='viridis', fmt='.2f', cbar=True, ax=axes[1])
axes[1].set_title('Income_Category vs. Diabetes_012')
axes[1].set_xlabel('Diabetes_012')
axes[1].set_ylabel('Income Category')

plt.show()
```

```python
cols_to_remove=['Diabetes_012','CholCheck','HeartDiseaseorAttack','HvyAlcoholConsump','DiffWalk','NoDocbcCost','Education_Category','Income_Category']
X=dataset.drop(columns=cols_to_remove)
y=dataset['Diabetes_012']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=28)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_params={
    'decision_tree': {'model': DecisionTreeClassifier(random_state=28), 'params' :{'criterion' : ["gini", "entropy", "log_loss"]}
                     },
    'logistic_regression': {'model': LogisticRegression(random_state=28), 'params': {'solver' : ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'], 'C': [1,10,20]}
                           },
    'random_forest' : {'model': RandomForestClassifier(random_state=28), 'params': {'criterion' : ["gini", "entropy", "log_loss"], 'n_estimators': [10,40,70,100,120,150]}
                      },
    'kneighborsclassifier': {'model': KNeighborsClassifier(), 'params': {'n_neighbors':[i for i in range(1,11)]}
                            },
    'xgboost' : {'model': XGBClassifier(random_state=28),'params':{'n_estimators':[100,200,300,400],'max_depth':[i for i in range(4,24)]}
                }
}
```

```python
from sklearn.model_selection import GridSearchCV
scores=[]
for model,params in model_params.items():
    clf=GridSearchCV(params['model'],params['params'],cv=2,return_train_score=False)
    clf.fit(X_train,y_train)
    print({'model':model,
                   'best_params':clf.best_params_,
                   'best_score':clf.best_score_
                  })
    scores.append({'model':model,
                   'best_params':clf.best_params_,
                   'best_score':clf.best_score_
                  })
```

```python
results=pd.DataFrame(scores)
results.sort_values(by='best_score',ascending=False,inplace=True)
results
```
